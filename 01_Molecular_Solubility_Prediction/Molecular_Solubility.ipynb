# **My First ML Project**

## **Load data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

## **Data Preparation**

### data separation

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

### data splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

## **Model Building**

### **Linear Regression**

#### ***Training Model***

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### **Applying the Model to Make a Prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

#### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

lr_result = pd.DataFrame({
    'Method': ['LinearRegression'],
    'Training MSE': [lr_train_mse],
    'Training R2': [lr_train_r2],
    'Testing MSE': [lr_test_mse],
    'Testing R2': [lr_test_r2]
})

lr_result = lr_result.round(3)
lr_result
```

### **Random Forest**

#### ***Training Model***

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

#### **Applying the Model to Make a Prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
y_rf_train_pred
```

```python
y_rf_test_pred
```

#### **Evaluate Model Performance**

```python
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

rf_result = pd.DataFrame({
    'Method': ['Random Forest'],
    'Training MSE': [rf_train_mse],
    'Training R2': [rf_train_r2],
    'Testing MSE': [rf_test_mse],
    'Testing R2': [rf_test_r2]
})


rf_result = rf_result.round(3)
rf_result
```

### **Model Comparison**

```python
df_models = pd.concat([lr_result, rf_result], axis=0).reset_index(drop=True)
df_models = df_models.round(3)
df_models
```

#### **Data Visualization of Prediction  results**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#7CAE00" ,alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train,p(y_train),"#F8766D")
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_rf_train_pred, c="#7CAE00" ,alpha=0.3)

z = np.polyfit(y_train, y_rf_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train,p(y_train),"#F8766D")
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
```

### **Hyperparameter Tuning for Random Forest using Randomized Search**

As you suggested, let's use `RandomizedSearchCV` to efficiently tune the hyperparameters `max_depth` and `n_estimators`. This will randomly sample from the parameter space, often yielding good results much faster than an exhaustive grid search.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint

# Define the parameter distributions to sample from
param_dist = {
    'max_depth': randint(5, 12),
    'n_estimators': randint(150, 400),
    'min_samples_split': randint(10, 30),
    'min_samples_leaf': randint(5, 15)
}

# Initialize the Random Forest Regressor
rf_random = RandomForestRegressor(random_state=100)

# Initialize RandomizedSearchCV
# n_iter controls how many different parameter combinations are tried
random_search = RandomizedSearchCV(estimator=rf_random, param_distributions=param_dist,
                                   n_iter=50, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, random_state=100)

# Fit RandomizedSearchCV to the training data
random_search.fit(x_train, y_train)

# Print the best parameters and best score
print(f"Best Parameters: {random_search.best_params_}")
print(f"Best Score (Negative MSE): {random_search.best_score_}")
```

#### ***Training Tuned Random Forest Model (from Randomized Search)***

```python
# Get the best model from random search
best_rf_random_model = random_search.best_estimator_

# Make predictions with the best model
y_rf_random_train_pred = best_rf_random_model.predict(x_train)
y_rf_random_test_pred = best_rf_random_model.predict(x_test)
```

#### **Evaluate Tuned Model Performance (Randomized Search)**

```python
rf_random_train_mse = mean_squared_error(y_train, y_rf_random_train_pred)
rf_random_train_r2 = r2_score(y_train, y_rf_random_train_pred)

rf_random_test_mse = mean_squared_error(y_test, y_rf_random_test_pred)
rf_random_test_r2 = r2_score(y_test, y_rf_random_test_pred)

rf_random_result = pd.DataFrame({
    'Method': ['Random Forest (Randomized Tuned)'],
    'Training MSE': [rf_random_train_mse],
    'Training R2': [rf_random_train_r2],
    'Testing MSE': [rf_random_test_mse],
    'Testing R2': [rf_random_test_r2]
})

# --- TAMBAHIN BARIS INI ---
rf_random_result = rf_random_result.round(3)
# --------------------------

rf_random_result
```

### **Updated Model Comparison (Including Randomized Search Tuned RF)**

```python
df_models_final = pd.concat([lr_result, rf_result, rf_random_result], axis=0).reset_index(drop=True)
df_models_final = df_models_final.round(3)
display(df_models_final)
```
