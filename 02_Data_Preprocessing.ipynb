Handling Categorical data

Categroical data encoding with pandas

create data frame

```python
import pandas as pd
df = pd.DataFrame([['green', 'M', 10.1, 'class1'],
                   ['red', 'L', 13.5, 'class2'],
                   ['blue', 'XL', 15.3, 'class1']])

df
```

labeling columns names using the `.columns` function

```python
df.columns = ['color', 'size', 'price', 'classlabel']
df
```

mapping ordinal feature

```python
size_mapping = {'XL': 3,
                'L' : 2,
                'M' : 1}

size_mapping
```

```python
df['size'] = df['size'].map(size_mapping)
df
```

encoding class labels

mapping categorical labels to values

```python
import numpy as np

class_mapping = {label: idx for idx, label in
                 enumerate(np.unique(df['classlabel']))}
class_mapping
```

```python
df['classlabel'] = df['classlabel'].map(class_mapping)
df
```

reverse the mapping of values back to their original categoriacl labels.

```python
inverse_class_mapping = {idx: label for label, idx in class_mapping.items()}
df['classlabel'] = df['classlabel'].map(inverse_class_mapping)
df
```

performing one-hot encoding on nominal feature

```python
from sklearn.preprocessing import LabelEncoder

x = df[['color', 'size', 'price']].values
color_le = LabelEncoder()
x[:, 0] = color_le.fit_transform(x[:, 0])
x
```

Load Wine data from the UCI ML Repo

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
df_wine
```

Relabel the column name

```python
df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium',
                   'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                   'Proanthocyanins','Color intensity', 'Hue',
                   'OD280/OD315 of diluted wines','Proline']
df_wine
```

seperate X and Y from df

```python
from sklearn.model_selection import train_test_split

x, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

x, y
```

Perform the Actual data splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.3,
                                                    random_state=0,
                                                    stratify=y)
x_train, x_test, y_train, y_test
```
